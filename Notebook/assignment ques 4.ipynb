```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data = pd.read_csv('practice_data.1.csv')
data.head()
```

```python
import statsmodels.api as sm
x =data['Days Present']
y=data['Task Completion Rates']

x_constant=sm.add_constant(x)
model=sm.OLS(y,x_constant).fit()
print(model.summary())
```
